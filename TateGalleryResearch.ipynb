```python
import pandas as pd
import seaborn as sns
```

```python
artist_data= pd.read_csv ("C:/Users/xx/Desktop/artist_data.csv", delimiter=",", encoding='utf-8')
art_work= pd.read_csv ("C:/Users/xx/Desktop/artwork_data.csv", delimiter=",", encoding='utf-8')    
```

```python
artist_data
```

```python
artist_data = artist_data.rename(columns={'name': 'artist'})
artist_data = artist_data.rename(columns={'id': 'artist_id'})
artist_data = artist_data.rename(columns={'url': 'artist_url'})
artist_data
```

```python
art_work = art_work.rename(columns={'id': 'artwork_id'})
art_work = art_work.rename(columns={'url': 'artwork_url'})
art_work
```

```python
outer_merged = pd.merge(artist_data, art_work, on=['artist'])
outer_merged.shape 
```

```python
outer_merged
```

```python
outer_merged.hist(figsize=(20,20))
```

```python
#total description of correations
outer_merged.describe(include='all')
```

```python
#defining  female artists who died between 1990 and 1995
outer_merged.loc[(outer_merged['yearOfDeath'] >= 1990)
                      & (outer_merged['yearOfDeath'] <= 1995), :].sort_values(
                        by='gender', ascending=True).head(100)
```

```python
#defining  male artists who died between 1990 and 1995
outer_merged.loc[(outer_merged['yearOfDeath'] >= 1990)
                      & (outer_merged['yearOfDeath'] <= 1995), :].sort_values(
                        by='gender', ascending=False).head(100)
```

```python
#defining the artists who died in 1930
outer_merged[outer_merged['yearOfDeath'] == 1930]
```

```python
#gender correlation
df = outer_merged
df
```

```python
#gender correlation
top = pd.crosstab(df['gender'], df['gender'], margins=True)
prop = top.loc[:, 'Female'] / top.loc[:, 'All']
prop
```

```python
#shows female artists obviously 1000 rows - if ask for 'all' than shows NaN gender artists as the database is not properly cleaned 
df.sort_values(by='gender', ascending=True).head (1000)
```

```python
#shows male artists - obviously 1000 rows
df.sort_values(by='gender', ascending=False).head (1000)
```

```python
#shows correlation between gender and acquisition year
import matplotlib.pyplot as plt
%matplotlib inline
df.plot( x="gender", y=["acquisitionYear"])
plt.show()
```

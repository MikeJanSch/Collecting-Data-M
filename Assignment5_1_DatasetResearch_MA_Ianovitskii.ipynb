```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#dataset of Russian and English speaking propaganda tweet bots
tweetmedia1= pd.read_csv ("C:/Users/xx/Desktop/fivethirtyeight-russian-troll-tweets/IRAhandle_tweets_1.csv", delimiter=",", encoding='utf-8')
```

```python
tweetmedia1
```

```python
#defining  authors who have more than 500 followers, but less than 1000
tweetmedia1.loc[(tweetmedia1['followers'] >= 500) & (tweetmedia1['followers'] <= 1000), :].sort_values(
                        by='author', ascending=True).head(1000)
```

```python
#defining  authors who have tweeted after 2010 in English
tweetmedia1.loc[(tweetmedia1['publish_date'] >= '2010') & (tweetmedia1['language']=='English'), :].sort_values(
                       by='author',ascending=False).head(1000)
```

```python
df=tweetmedia1
```

```python
# defining the 5000 most followed tweets by BLMSOLDIER - showed 3579 tweets (all matching by author)
BLMSOLDIER = df.loc[df['author'] == 'BLMSOLDIER', :].nlargest(5000, 'followers')
BLMSOLDIER
```

```python
# defining the 1000 most retweeted tweet by 10_GOP - showed 372 tweets (all matching by author)
TEN_GOP = df.loc[df['author'] == '10_GOP', :].nlargest(1000, 'retweet')
TEN_GOP
```

```python
#filtering language of tweets as 'English' 
language_filter = df['language'] == 'English'
language_filter
```

```python
#defining tweets in Russian 
language_filter = df['language'] == 'Russian'
russian_df = df[language_filter]
russian_df.sort_values(by='author',ascending=False).head(1000)
```

```python
#defining the number of updates of tweets made by authors
df_author_by_updates = df.groupby('author')['updates'].sum()
df_author_by_updates
```

```python
#defining 30 most updated authors
df_author_by_updates.sort_values(ascending=False)[:30]
```

```python
#visualize top 30 authors by updates
top30_authors = df_author_by_updates.sort_values(ascending=False)[:30]
top30_authors.plot(kind='bar')
```

```python
#visualize top 30 authors by updates and save picture
ax = top30_authors.plot(kind='bar')
plt.tight_layout()
ax.figure.savefig('top30_authors.png')
```

```python
#defining authors with more than 1000 followers
update_df = df.drop(df[df['followers'] > 1000].index)
update_df
```

```python
#visualizing and defining top 20 authors by number of persons he/she is following
df_author_by_following = df.groupby('author')['following'].sum()
df_author_by_following.sort_values(ascending=False)[:20]
top20_authors = df_author_by_updates.sort_values(ascending=False)[:20]
top20_authors.plot(kind='bar')
```

```python
#Defining correlation between numbers of followers and following of 'ATLANTA_ONLINE''s tweets
BotAtlanta = df.loc[df['author']== 'ATLANTA_ONLINE', :]
sns.lmplot(x='following', y='followers', data=BotAtlanta)
```

```python
#visualizing and defining the correlation between followers and following in general
ax = df.plot(kind='scatter', x='following', y='followers', color='blue')
```
